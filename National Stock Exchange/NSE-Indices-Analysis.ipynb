```python
import os
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
import numpy as np
matplotlib.rcParams['figure.figsize'] = (20,6)

os.listdir('/kaggle/input')


def bar_color(df,color1,color2):
    return np.where(df.values>0,color1,color2).T


data = pd.read_csv('/kaggle/input/nse-indices-data-fy22-23/MW-All-Indices-30-May-2023.csv')

row_data = data['365 D %']
row_data1 = data['30 D %']
name = data['Index']

xpos = np.arange(len(name))

plt.xticks(xpos, name, rotation='vertical')

plt.ylabel('% Change')
plt.xlabel('NIFTY indices')

plt.bar(xpos-0.2, row_data, width=0.4,color=bar_color(row_data, '#20d420', 'r'), label='May 2022-23')
plt.bar(xpos+0.2, row_data1, width=0.4, label='Apr-May 2023', color=bar_color(row_data1, '#f99820', '#2081f9'))
plt.grid(axis='x', alpha=0.25)
plt.grid(axis='y', alpha=0.50)


plt.legend(loc='upper left')
plt.show()
```
